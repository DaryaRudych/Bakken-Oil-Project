```python
#Importing dependencies 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#reading file into a dataframe
file = "clean_data.csv"
df = pd.read_csv(file)
df.columns
```

```python
#Overview of Eagle Ford Play
summary = df.groupby(['Sub-play'])['Operator', "API"].nunique()
summary = summary.rename(columns = {"API": "Well count", "Operator": "Operator Count"})
```

```python
# Plot activity by SUb-play
with plt.style.context('fivethirtyeight'):
    summary['Operator Count'].plot(kind='bar',figsize=(15, 7),rot=70)

    # Add axis labels and title
    plt.title("Operator Count by Sub-play")
    plt.ylabel("Operator Count")
    

    # Show plot
    plt.show()
```

```python
from matplotlib import pyplot

#Set the colors
colors = ['black', 'red', 'green', 'blue', 'cyan', 'yellow', 'orange', 'brown', 'magenta', 'salmon']

#Create a list of ticks
x = []
subplays = list(df['Sub-play'].unique())
ticks = 0
for sp in df['Sub-play'].unique():
    ticks = ticks + 1
    x.append(ticks)
    
    
y1 = summary['Well count'].sort_values(ascending=False)
y2 = summary['Operator Count'].sort_values(ascending=False)
with plt.style.context('fivethirtyeight'):
    f, (ax1, ax2) = plt.subplots(1, 2, sharex=True)
    f.set_figheight(25)
    f.set_figwidth(55)
    plt.subplot(1, 2, 1)
    chart1 = plt.bar(x, y2, align='center', alpha=0.5, color = colors)
    plt.ylabel('Operator count', fontsize=30)
    plt.xticks(x, fontsize = 30)
    plt.yticks(fontsize = 20)
plt.legend(chart1, subplays)

with plt.style.context('fivethirtyeight'):
    plt.suptitle('Activity by sub-play', fontsize=40)
    plt.subplot(1, 2, 2)
    chart2 = plt.bar(x, y1, align='center', alpha=0.5, color = colors)
    plt.ylabel('Well count', fontsize=30)
    plt.subplots_adjust(left  = 0.2, right = 0.9, bottom = 0.3, top = 0.9, wspace = 0.07, hspace = 0.6 )
    plt.xticks(x, fontsize = 30)
    plt.yticks(fontsize = 20)

plt.legend(chart2, subplays)


plt.savefig('ActivityBySubPlay.png', format='png', dpi=96, transparent=True)
plt.show()
```

```python
#Overview of Eagle Ford Play

#Calculating summary statistics for each sub-play

agg = {
    'Well Cost Total (US$)': {'Avg Well cost': 'mean'}, 
    'Rig Cost (US$)': {'Avg Rig Cost (US$)':'mean'}, 
    'EUR Total (mmboe)':{'Avg EUR (mmboe)': 'mean'},
    'Cum365 Gas (mcf)': {'Avg Cum365 Gas (mcf)': 'mean', 'Max Cum365 Gas (mcf)': max},
    'Cum365 Oil (bbl)': {'Avg Cum365 Oil (bbl)': 'mean', 'Max Cum365 Oil (bbl)': max}

}


summary_1 = df.groupby(['Sub-play']).agg(agg)
summary_1.columns = summary_1.columns.droplevel()
summary_1
```

```python
#Merging two dataframs into one to get an overview of Eagle Ford Play by Sub-play:
summary_final  = pd.merge(summary, summary_1, left_index=True, right_index=True)
summary_final
```

```python
# data to plot
n_groups = 10
sub_plays = summary_final.index[0:10]


# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.4
opacity = 0.8
 
rects1 = plt.bar(index, summary_final['Avg Cum365 Gas (mcf)'], bar_width, alpha=opacity,
                 color='b', align='center',
                 label='Avg Cum365 Gas (mcf)')
 
rects2 = plt.bar(index + bar_width, summary_final['Avg Cum365 Oil (bbl)'] , bar_width,
                 alpha=opacity,
                 color='g',
                 label='Avg Cum365 Oil (bbl)', 
                 align='center')
 
plt.xlabel('Sub-plays')
plt.ylabel('Avg Production of oil&gas')
plt.title('Comparison of oil/gas production by sub-play')
plt.xticks(index, sub_plays, rotation=50, rotation_mode="anchor", horizontalalignment="right")

plt.legend(loc=2)
 
plt.tight_layout()
fig.set_size_inches(19,8)
plt.show()
```
